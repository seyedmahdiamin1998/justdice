import necessary libraries

```python
import pandas as pd
import numpy as np

import plotly.express as px
```

```python
# adspend
df_adspend = pd.read_csv('../../data/adspend.csv', parse_dates=['event_date'])
df_adspend['country_id'] = df_adspend.country_id.astype('category')
df_adspend['network_id'] = df_adspend.network_id.astype('category')
df_adspend['client_id'] = df_adspend.client_id.astype('category')

# installs
df_installs = pd.read_csv('../../data/installs.csv', parse_dates=['event_date'])
df_installs['country_id'] = df_installs.country_id.astype('category')
df_installs['app_id'] = df_installs.app_id.astype('category')
df_installs['network_id'] = df_installs.network_id.astype('category')
df_installs['device_os_version'] = df_installs.device_os_version.astype('category')
df_installs.drop_duplicates(inplace=True)

# payouts
df_payouts = pd.read_csv('../../data/payouts.csv', parse_dates=['event_date'])
df_payouts = df_payouts.groupby(by=['install_id'])['value_usd'].sum().reset_index(name='payouts_usd').sort_index()

# revenue
df_revenue = pd.read_csv('../../data/revenue.csv', parse_dates=['event_date'])
df_revenue = df_revenue.groupby(by=['install_id'])['value_usd'].sum().reset_index(name='revenue_usd').sort_index()
```

___

```python
df = df_installs.merge(df_payouts, on='install_id', how='inner').merge(df_revenue, on='install_id', how='inner')
df.head()
```

```python
vis_df = df[['event_date', 'payouts_usd', 'revenue_usd']].\
            groupby(by=['event_date']).\
            sum().\
            reset_index().\
            sort_values(by=['event_date'])

fig = px.line(vis_df,\
            x="event_date", 
            y=vis_df.columns, 
            template="plotly_white",
            color_discrete_sequence=['#131148', '#00ba58'])

fig.add_layout_image(
    dict(
        source="../../assets/just_dice.svg",
        xref="paper", yref="paper",
        x=0, y=1.03,
        sizex=0.2, sizey=0.2,
        xanchor="left", yanchor="bottom"
    )
)
fig.show()
del vis_df
```

```python
vis_df = df[['network_id', 'payouts_usd', 'revenue_usd']].\
            groupby(by=['network_id']).\
            sum().\
            reset_index().\
            sort_values(by=['network_id'])


vis_df['network_id'] = vis_df['network_id'].astype('category')
vis_df['network_id'] = vis_df['network_id'].apply(lambda x : f'network {x}')

vis_df = vis_df.melt(id_vars=['network_id'], value_vars=['payouts_usd','revenue_usd'])
vis_df.rename(columns={'variable':'type'}, inplace=True)
fig = px.bar(vis_df, 
            x='network_id', 
            y='value', 
            color='type', 
            barmode='group', 
            template='plotly_white',
            color_discrete_sequence=['#131148', '#00ba58'])


fig.update_layout(
    yaxis_title='value $',
    xaxis_title='Network'
)

fig.add_layout_image(
    dict(
        source="../../assets/just_dice.svg",
        xref="paper", yref="paper",
        x=0, y=1.03,
        sizex=0.2, sizey=0.2,
        xanchor="left", yanchor="bottom"
    )
)

fig.show()

del vis_df
```
